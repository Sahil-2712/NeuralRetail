```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
from pathlib import Path
```

```python
# The raw URL for your specific file
url = "https://github.com/IamArunavaSamanta/NeuralRetailProject/raw/main/data/raw/online_retail_II.xlsx"

# To read a specific sheet (e.g., 'Year 2009-2010')
data1 = pd.read_excel(url, sheet_name='Year 2009-2010', engine='openpyxl')
data2 = pd.read_excel(url, sheet_name='Year 2010-2011', engine='openpyxl')
```

```python
# Append two data sheet
df = pd.concat([data1, data2], axis=0, ignore_index=True)
```

```python
df.info()
```

```python
print("Total number of columns are: ", len(df.columns))
print("Total number of rows are: ", len(df))
```

```python
initial_len = len(df)
df.drop_duplicates(inplace=True)
final_len = len(df)

print(f"Removed {initial_len - final_len} duplicate rows.")
print(f"New length: {final_len}")
```

```python
# Rename Columns
df = df.rename(columns={'Customer ID': 'CustomerID', 'Invoice': 'InvoiceNo' })
```

```python
df['Description'] = df['Description'].fillna('Unknown')
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CustomerID'] = pd.to_numeric(df['CustomerID'], errors='coerce')
df['CustomerID'] = df['CustomerID'].apply(lambda x: str(int(x)) if pd.notna(x) else x)
```

```python
#Add new columns
df['Revenue'] = df['Quantity'] * df['Price']
```

```python
df.isnull().sum()
```

```python
# Remove rows with negative quantity or price
before = len(df)
df = df[(df["Quantity"] > 0) & (df["Price"] > 0)]
after = len(df)
print("Dropped", before - after, "rows with negative quantity or price")
```
